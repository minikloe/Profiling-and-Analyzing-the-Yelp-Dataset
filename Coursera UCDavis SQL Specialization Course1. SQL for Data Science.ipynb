illusion table, stored query, simplify querires

`AS` function


https://zhuanlan.zhihu.com/p/80183774  
视图View, 基于SQL语句的结果集 的可视化的表, 包含行和列,是一种不存在的虚拟表.  
视图中的字段是来自一个或多个数据库中的真实的表中的字段  
可以向视图添加 SQL 函数、WHERE 以及 JOIN 语句，也可以提交数据，就像这些来自于某个单一的表。  
数据库的设计和结构不会受到视图中的函数、where 或 join 语句的影响。  
视图总是显示最近的数据。每当用户查询视图时，数据库引擎通过使用 SQL 语句来重建数据。  


#### 3. SQL CREATE VIEW 实例

SQL CREATE VIEW 语法：
```sql
CREATE VIEW view_name AS
SELECT column_name(s)
FROM table_name
WHERE condition

--创建视图--
create or replace view v_student as 
select Sno,Sname
from student;

--从视图中检索数据，即查询上面这个视图--
select * from v_student;

--也可以向查询添加条件--
select * 
from v_student
where Sname like '%云‘;

--删除视图--
drop view v_student;

```
